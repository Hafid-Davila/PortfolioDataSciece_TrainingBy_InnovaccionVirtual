```python
# Conectamos con Azure

from azure.cognitiveservices.vision.face import FaceClient
from msrest.authentication import CognitiveServicesCredentials

client = FaceClient(endpoint, CognitiveServicesCredentials(key))
```

```python
# preparamos el entorno

import faces 
import os 
%matplotlib inline
```

```python
# Obetenemos una imagen para detectar rostros

image_path = './Data/VisionComputacional/moise-m-HTP5klp7fn4-unsplash.jpg'
image_stream = open(image_path, 'rb')

detected_faces = client.face.detect_with_stream(image = image_stream)

faces.show_faces(image_path, detected_faces)
```

```python
# Obetenemos una imagen para detectar rostros y su identificaci√≥n

image_path = './Data/VisionComputacional/moise-m-HTP5klp7fn4-unsplash.jpg'
image_stream = open(image_path, 'rb')

detected_faces = client.face.detect_with_stream(image = image_stream)

faces.show_faces(image_path, detected_faces, show_id=True)

```

```python
# Obetenemos una imagen para dectectar rostros y sus atributos

image_path = './data/visioncomputacional/frederik-trovatten-com-8ikTPI-novk-unsplash.jpg'
image_stream = open(image_path, 'rb')

attributes = ['age', 'emotion']
detected_faces = client.face.detect_with_stream(image = image_stream, return_face_attributes=attributes)

faces.show_face_attributes(image_path, detected_faces)
```
